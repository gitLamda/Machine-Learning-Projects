```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import sklearn.datasets
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the data into a data frame
df = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
df.head()
```

```python
#adding the target column to the data frame
df['label'] = breast_cancer_dataset.target
```

```python
df.head()
```

```python
df.tail()
```

```python
df.groupby('label').mean()
```

```python
df['label'].value_counts()
```

```python
X = df.drop('label',axis = 1)
y = df['label']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, stratify = y, random_state = 42)
```

# **Building the Nural Network**

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of Nural Network
#Input shape is equal to the number of features in the data set. Flatten converts X_train matrix into single dimentional array
#2 and 20 are the number of nurons. 2 is given as output is because we only have two different classes. if you have three input 3 nurons
model = keras.Sequential ([
                          keras.layers.Flatten(input_shape = (30,)),
                          keras.layers.Dense(20,activation='relu'),
                          keras.layers.Dense(2,activation='sigmoid')
])
```

```python
#compiling the nural network
#when we have one hot encoding we use categorical_crossentropy
model.compile(optimizer ='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
#training the nural network
history = model.fit(X_train, y_train, validation_split=0.1,epochs = 10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')

plt.legend(['Training Data','Validation Data'],loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(loss)
print(accuracy)
```

```python
y_pred = model.predict(X_test)
```
