# **Machine Learning Model**

```python
X=sonar_data.drop(60,axis=1)
y=sonar_data[60]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(X_test)
print(y_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
#Checking for accuracy of Test data
X_test_prediction=model.predict(X_test)
accuracy_testing=accuracy_score(X_test_prediction,y_test)

print(f"Accuracy Score for Test data: {accuracy_testing*100}")
```

```python
#Checking for accuracy of Train data
X_train_prediction=model.predict(X_train)
accuracy_training=accuracy_score(X_train_prediction,y_train)

print(f"Accuracy Score for Train data: {accuracy_training*100}")
```
