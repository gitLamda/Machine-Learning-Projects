```python
#Seprating data and labels
X=wine_data.drop(['quality'],axis=1)
```

```python
#Label Binarization
y=wine_data['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
#accuracy score for test data
X_test_prediction=model.predict(X_test)
accuracy_testing=accuracy_score(X_test_prediction,y_test)
print(f'Accuracy for test data: {accuracy_testing*100}')
```

```python
#accuracy score for train data
X_train_prediction=model.predict(X_train)
accuracy_training=accuracy_score(X_train_prediction,y_train)
print(f'Accuracy for test data: {accuracy_training*100}')
```

```python
input_data = (7.8,0.58,0.02,2,0.073,9,18,0.9968,3.36,0.57,9.5)

# changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the data as we are predicting the label for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if prediction[0]==0:
  print('Bad Quality')
else:
  print('Good Quality')
```
